```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('datasets/Retail_Store_Cleaned_Consumer.csv', encoding='latin-1')
```

```python
df.shape
```

### Delivery Performance

Pie Chart of Delivery Status

```python
status_counts = df['Delivery Status'].value_counts()

# To assign Late delivery as the seaborn palette red
color = sns.color_palette("Set2")
palette = sns.color_palette([color[1], color[0], color[2]])

plt.figure(figsize=(6, 6))
plt.pie(status_counts, 
        labels=status_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=palette)
plt.title("Proportion of Delivery Status")
plt.axis('equal')
plt.show()
```

Analysing Shipping Mode
* First Class
* Standard Class
* Second Class
* Same day

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Delivery Status', hue='Shipping Mode', data=df, palette='Set2')

plt.title('Delivery Status by Shipping Mode', fontsize=14)
plt.xlabel('Delivery Status')
plt.ylabel('Number of Deliveries')
plt.show()
```

Late Delivery Risk by Order Region

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='Late_delivery_risk', hue='Order Region', palette='coolwarm')
plt.title('Late Delivery Risk by Order Region', fontsize=14)
plt.xlabel('Order Region', fontsize=12)
plt.ylabel('Number of Deliveries', fontsize=12)
# plt.legend(title='Late Delivery Risk', fontsize=10)
plt.show()
```
