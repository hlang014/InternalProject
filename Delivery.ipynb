### Delivery Performance

Pie Chart of Delivery Status

```python
status_counts = df['Delivery Status'].value_counts()

# Assigning colors for each status
colors = {
    'Late delivery': '#FFA500',  
    'Shipping on time': '#66b366',  
    'Advance shipping': '#99ff99'
}  

status_colors = [colors.get(status, '#cccccc') for status in status_counts.index]

plt.figure(figsize=(8,8))
plt.pie(status_counts, labels = status_counts.index, autopct='%1.1f%%', startangle=90,colors=status_colors)
plt.title("Proportion of Delivery status")
plt.axis('equal')
plt.show()
```
